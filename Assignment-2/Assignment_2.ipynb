2. Removing Punctuation Marks
3. Stemming
4. Lemmatization

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
import re
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
para1 = (
    "Artificial intelligence is transforming every industry! From healthcare to finance, "
    "it has revolutionized the way we work; and live. Are machines replacing humans? "
    "Some say yes... others disagree: both perspectives have merit, don't they?"
)

para2 = (
    "Climate change is one of the most pressing challenges of our time!!! "
    "Scientists warn: rising temperatures, melting glaciers; and extreme weather events "
    "are increasing at an alarming rate. Can we still reverse the damage? "
    "Many experts believe so... but immediate action is required."
)

text = para1 + " " + para2
print("=== Original Combined Text ===")
print(text)
```

## Step 1: Removing Punctuation Marks *(Extra Credit)*

```python
punct_removed = re.sub(r'[' + re.escape(string.punctuation) + r']+', ' ', text)
punct_removed = ' '.join(punct_removed.split())

print("=== After Punctuation Removal ===")
print(punct_removed)
```

## Step 2: Stop Word Removal

```python
stop_words = set(stopwords.words('english'))
tokens = word_tokenize(punct_removed.lower())
tokens_no_stop = [w for w in tokens if w.isalpha() and w not in stop_words]

print("=== After Stop Word Removal ===")
print(tokens_no_stop)
```

## Step 3: Stemming

```python
ps = PorterStemmer()
stemmed = [ps.stem(w) for w in tokens_no_stop]

print("=== After Stemming ===")
print(stemmed)
```

## Step 4: Lemmatization

```python
wnl = WordNetLemmatizer()
lemmatized = [wnl.lemmatize(w) for w in tokens_no_stop]

print("=== After Lemmatization ===")
print(lemmatized)
```

```python
final_text = ' '.join(lemmatized)

print("=== Final Clean Text ===")
print(final_text)
```

# Part 2: Feature Engineering

## Bag of Words

```python
%pip install -q scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np

clean_para1 = [
    wnl.lemmatize(w)
    for w in word_tokenize(re.sub(r'[' + re.escape(string.punctuation) + r']+', ' ', para1).lower())
    if w.isalpha() and w not in stop_words
]

clean_para2 = [
    wnl.lemmatize(w)
    for w in word_tokenize(re.sub(r'[' + re.escape(string.punctuation) + r']+', ' ', para2).lower())
    if w.isalpha() and w not in stop_words
]

corpus = [
    ' '.join(clean_para1),
    ' '.join(clean_para2),
    final_text
]

bow_vectorizer = CountVectorizer(max_features=60)
bow_matrix = bow_vectorizer.fit_transform(corpus)
bow_array = bow_matrix.toarray()
feature_names = bow_vectorizer.get_feature_names_out()

print('=== Bag of Words ===')
print('Vocabulary size:', len(feature_names))
print('Vocabulary sample (first 20):', list(feature_names[:20]))
print('\nBOW matrix shape:', bow_array.shape)
print('Rows: para1_clean, para2_clean, combined_clean')
print('\nBOW matrix (first 20 columns):')
print(bow_array[:, :20])
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=50)
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
tfidf_array = tfidf_matrix.toarray()

feature_names_tfidf = tfidf_vectorizer.get_feature_names_out()

print("=== TF-IDF ===")
print("Word vocabulary:", list(feature_names_tfidf))
print("\nTF-IDF Vector (first 20 features):")
print(np.round(tfidf_array[0][:20], 4))
print("\nTF-IDF shape:", tfidf_array.shape)
```

# Part 3: Word2Vec

Word embeddings with more than 200 words input and similarity scoring

```python
import gensim
print('gensim version:', gensim.__version__)
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import sent_tokenize

w2v_text = """
Artificial intelligence systems are changing education, healthcare, banking, transportation, and agriculture in remarkable ways.
Students now use adaptive learning platforms that personalize lessons, detect weak concepts, and recommend practice material.
Teachers use analytics dashboards to identify learning gaps, design focused interventions, and monitor progress over time.
Hospitals use machine learning for risk prediction, medical imaging support, triage automation, and patient flow optimization.
Doctors still make final decisions, but data-driven tools improve speed, consistency, and confidence in diagnosis.
In finance, fraud detection models inspect transaction patterns, geolocation signals, and behavior profiles to flag suspicious activity.
Banks also apply natural language processing to classify support tickets, summarize conversations, and route customer issues.
Supply chain teams forecast demand using seasonal trends, promotion effects, macroeconomic indicators, and real-time inventory updates.
Manufacturing plants monitor vibration, temperature, pressure, and power consumption to predict equipment failures early.
Predictive maintenance reduces downtime, lowers repair costs, and improves safety for engineers working near heavy machinery.
Retail companies build recommendation engines that connect users to relevant products based on browsing and purchase history.
Marketing teams test campaign messages, compare conversion rates, and optimize audience targeting with controlled experiments.
Autonomous vehicles combine cameras, radar, and lidar to interpret roads, detect obstacles, and plan safe trajectories.
Robotics platforms use computer vision for object recognition, pose estimation, grasp planning, and quality inspection tasks.
Cloud platforms provide scalable infrastructure so teams can train large models without buying expensive local hardware.
Engineers track metrics such as precision, recall, latency, and throughput to evaluate both quality and deployment efficiency.
Model monitoring is essential because data drift, concept drift, and seasonality can degrade performance after deployment.
Responsible AI practices include fairness checks, transparency reports, privacy protection, and clear accountability policies.
Organizations train staff in data literacy so decisions are guided by evidence instead of assumptions or anecdotes.
Cross-functional collaboration between product managers, domain experts, analysts, and engineers improves project outcomes.
Open-source libraries accelerate experimentation, but teams must validate assumptions with domain-specific datasets and constraints.
Strong documentation helps future contributors reproduce results, compare baselines, and understand design tradeoffs.
Good communication of uncertainty prevents overconfidence and supports better decisions in high-impact environments.
Continuous improvement cycles keep systems reliable as business goals, user behavior, and regulations evolve.
"""

sentences = []
for sentence in sent_tokenize(w2v_text):
    tokens = [token.lower() for token in word_tokenize(sentence) if token.isalpha()]
    if tokens:
        sentences.append(tokens)

total_words = sum(len(s) for s in sentences)
print('=== Word2Vec Input Check ===')
print('Sentence count:', len(sentences))
print('Word count:', total_words)
if total_words <= 200:
    raise ValueError('Input must contain more than 200 words for this assignment.')

w2v_model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1,
    seed=42
)

print('\n=== Word2Vec Model ===')
print('Vocabulary size:', len(w2v_model.wv))
print('Vector dimensions:', w2v_model.vector_size)
print("Sample vector for 'learning' (first 10 values):")
print(w2v_model.wv['learning'][:10])
```

## Word Similarity Scores

```python
word_pairs = [
    ('learning', 'education'),
    ('model', 'prediction'),
    ('finance', 'banking'),
    ('robotics', 'vision'),
    ('healthcare', 'hospital'),
    ('data', 'analytics'),
    ('monitoring', 'drift'),
    ('fairness', 'privacy')
]

print('=== Word Similarity Scores using Cosine Similarity ===\n')
print(f"{'Word 1':<15} {'Word 2':<15} {'Similarity Score':<20}")
print('-' * 55)

for word1, word2 in word_pairs:
    if word1 in w2v_model.wv and word2 in w2v_model.wv:
        similarity = w2v_model.wv.similarity(word1, word2)
        print(f"{word1:<15} {word2:<15} {similarity:.4f}")
    else:
        print(f"{word1:<15} {word2:<15} Not in vocabulary")

print("\n=== Most Similar Words to 'learning' ===")
for word, score in w2v_model.wv.most_similar('learning', topn=5):
    print(f"{word}: {score:.4f}")

print("\n=== Most Similar Words to 'model' ===")
for word, score in w2v_model.wv.most_similar('model', topn=5):
    print(f"{word}: {score:.4f}")
```

## Part 1: Cosine Similarity Between 3 Sentences

Calculate similarity scores between sentence pairs using TF-IDF vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

test_sentences = [
    'Machine learning requires clean data and proper feature engineering.',
    'Deep learning models need abundant training data and computational resources.',
    'Natural language processing enables machines to understand human text and context.'
]

tfidf_vec = TfidfVectorizer(max_features=100)
tfidf_test_matrix = tfidf_vec.fit_transform(test_sentences)

sim_matrix = cosine_similarity(tfidf_test_matrix)

print('=== Cosine Similarity Between 3 Sentences ===\n')
print('Sentence 1:', test_sentences[0])
print('Sentence 2:', test_sentences[1])
print('Sentence 3:', test_sentences[2])

print('\n=== Similarity Matrix ===')
print('        Sent1  Sent2  Sent3')
for i in range(3):
    print(f'Sent{i+1}: ', end='')
    for j in range(3):
        print(f'{sim_matrix[i][j]:.4f} ', end='')
    print()

print('\n=== Pairwise Similarity Scores ===')
print(f'Sentence 1 vs Sentence 2: {sim_matrix[0][1]:.4f}')
print(f'Sentence 1 vs Sentence 3: {sim_matrix[0][2]:.4f}')
print(f'Sentence 2 vs Sentence 3: {sim_matrix[1][2]:.4f}')
```

## Part 2: Text Classification (Supervised Learning)

Classify sentences as "Technical" or "Non-Technical" using Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

training_texts = [
    'Python is a high level programming language',
    'Artificial neural networks process information through layers',
    'Convolutional networks are used for image recognition',
    'Machine learning algorithms learn patterns from data',
    'The weather today is sunny and warm',
    'Dogs are loyal and friendly animals',
    'Cats sleep most of the day indoors',
    'Birds migrate to warmer regions in winter'
]

training_labels = [
    1, 1, 1, 1,
    0, 0, 0, 0
]

nb_classifier = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=50)),
    ('nb', MultinomialNB())
])

nb_classifier.fit(training_texts, training_labels)

predictions = nb_classifier.predict(training_texts)
accuracy = accuracy_score(training_labels, predictions)

print('=== Text Classification with Naive Bayes ===')
print(f'Accuracy on training data: {accuracy:.4f}')

print('\n=== Predictions ===')
for i, text in enumerate(training_texts):
    pred = predictions[i]
    label = 'Technical' if pred == 1 else 'Non-Technical'
    print(f'{i+1}. "{text}" → {label}')

print('\n=== Confusion Matrix ===')
cm = confusion_matrix(training_labels, predictions)
print(cm)

print('\n=== Classification Report ===')
print(classification_report(training_labels, predictions, 
                          target_names=['Non-Technical', 'Technical']))
```

## Part 3: Text Clustering (Unsupervised Learning)

Group unlabeled text documents using K-Means clustering

```python
from sklearn.cluster import KMeans

clustering_texts = [
    'Python is a popular programming language for machine learning',
    'Java and C++ are used for building large scale applications',
    'JavaScript is essential for web development and frontend design',
    'Dogs are mammals that live on land and water environments',
    'Cats are independent animals that sleep and hunt',
    'Birds have wings and can fly across the sky',
    'Fish live in water and have gills for breathing',
    'Neural networks mimic biological systems for computation',
    'Algorithms are step by step procedures to solve problems',
    'Data structures organize information efficiently'
]

tfidf_cluster = TfidfVectorizer(max_features=50)
X_cluster = tfidf_cluster.fit_transform(clustering_texts)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(X_cluster)

print('=== K-Means Text Clustering ===')
print(f'Number of clusters: 3')
print(f'Number of documents: {len(clustering_texts)}')
print(f'Inertia (within-cluster sum of squares): {kmeans.inertia_:.4f}')

print('\n=== Cluster Assignments ===')
for cluster_id in range(3):
    print(f'\nCluster {cluster_id}:')
    cluster_docs = [clustering_texts[i] for i in range(len(clustering_texts)) 
                   if cluster_labels[i] == cluster_id]
    for idx, doc in enumerate(cluster_docs, 1):
        print(f'  {idx}. {doc}')

print('\n=== Raw Cluster Labels ===')
print('Document → Cluster ID')
for i, label in enumerate(cluster_labels):
    print(f'{i+1}. "{clustering_texts[i]}" → Cluster {label}')
```
